```python
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv('./data/train.csv')

df.head()
```

```python
df.info()
```

```python
sns.set(style="darkgrid")
sns.relplot(x="Age", y="Fare", data=df);
```

```python
sns.relplot(x="Age", y="Fare",hue="Sex", data=df);
```

```python
sns.relplot(x="Age", y="Fare",hue="Sex",style="Sex", data=df);
```

```python
sns.relplot(x="Age", y="Fare",hue="Sex",style="Survived", data=df);
```

```python
sns.relplot(x="Age", y="Fare",hue="Embarked",style="Survived", data=df);
```

```python
sns.relplot(x="Age", y="Fare",hue="Sex",style="Embarked", data=df);
```

```python
sns.relplot(x="Age", y="Fare",hue="Sex",size="Embarked", style="Survived", data=df);
```
